```python
# 1. Installation
!pip install PyPDF2 openai

import PyPDF2
from openai import OpenAI
from google.colab import files
try:
    from google.colab import userdata
    # Versucht den Key aus den Colab-Secrets zu lesen
    api_key = userdata.get('OPENAI_API_KEY')
except:
    # Falls nicht gesetzt, hier manuell eintragen oder Abfrage
    api_key = "DEIN_KEY_HIER_WENN_KEIN_SECRET_GENUTZT_WIRD"

# 2. Datei-Upload
print("Bitte lade dein PDF hoch:")
uploaded = files.upload()
if not uploaded:
    print("Keine Datei ausgewählt.")
else:
    file_name = list(uploaded.keys())[0]

    # 3. Text extrahieren
    def extract_text_from_pdf(pdf_path):
        reader = PyPDF2.PdfReader(pdf_path)
        text = ""
        for page in reader.pages:
            content = page.extract_text()
            if content:
                text += content
        return text

    pdf_text = extract_text_from_pdf(file_name)

    # 4. KI-Zusammenfassung
    client = OpenAI(api_key=api_key)

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "Summarize this document for a busy executive. Use bullet points."},
                {"role": "user", "content": pdf_text[:4000]}
            ]
        )

        print("\n--- ZUSAMMENFASSUNG ---")
        print(response.choices[0].message.content)
    except Exception as e:
        print(f"\nFehler: {e}")
        print("Hinweis: Überprüfe, ob dein OpenAI API-Key korrekt hinterlegt ist.")
```
